```python
# replacing 'present' with current date
from datetime import datetime
today_date = datetime.today().date()
df['DOJ'] = df['DOJ'].replace('present', today_date)
df['DOL'] = df['DOL'].replace('present', today_date)
```

```python
col = ['DOB', 'DOJ', 'DOL']

df[col] =df[col].apply(pd.to_datetime)
```

```python
df = df.drop(df[~(df['DOL'] > df['DOJ'])].index)
df.shape
```

```python
# replacing -1,0 values in few columns  with nan 
```

```python
selected_num_columns = ['Domain','ComputerProgramming', 'ElectronicsAndSemicon',
       'ComputerScience', 'MechanicalEngg', 'ElectricalEngg','TelecomEngg',
       'CivilEngg']
df[selected_num_columns] = df[selected_num_columns].replace(-1, np.nan)
```

```python
selected_cat_columns = ['10board','12board', 'GraduationYear']
       
df[selected_cat_columns] = df[selected_cat_columns].replace('0', np.nan)
```

```python
df['JobCity']=df['JobCity'].replace('-1',np.nan )
```

```python
# imputing nan values with mean, mode
```

```python
df[selected_num_columns] = df[selected_num_columns].fillna(df[selected_num_columns].mean())

```

```python
df[selected_cat_columns] = df[selected_cat_columns].fillna(df[selected_cat_columns].mode().iloc[0])
```

```python
df.dtypes
```

```python
cat_cols=['10board','12board', 'GraduationYear','Specialization','CollegeState','Designation','JobCity']
```

```python
# considering only top 15 categories in categorical columns
def group_categorical(df, categorical_columns):
    for col in categorical_columns:
        value_counts=df[col].value_counts()
        top_10_groups=value_counts.head(15).index
        other_groups=value_counts.index.difference(top_10_groups)
        df[col]=df[col].replace(other_groups,'Others')
    return df
```

```python
def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 *IQR)
    outliers_below=column[column<lower_bound]
    outliers_above=column[column>upper_bound]
    outliers_count=len(outliers_below)+len(outliers_above)
    return     lower_bound,upper_bound,len(outliers_below),len(outliers_above),outliers_count
```

```python
# finding outliers 
def find_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 *IQR)
    outliers=column[(column<lower_bound) |(column>upper_bound)]
    return outliers
    
```

## Research Questions
### Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
###  Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python
cse_data=df[df['Degree']=='B.Tech/B.E.']
job_roles=[ 'Programming Analyst', 'Software Engineer', 'Hardware Engineer' ,'Associate Engineer']
req_data=cse_data[cse_data['Designation'].isin(job_roles)]
salary_range=(250000,300000)
avg_sal=req_data['Salary'].mean()
if salary_range[0]<=avg_sal<=salary_range[1]:
    print('The claim is supported by the Data')
else:
    print('The claim is not supported by the Data')
```

```python
from scipy import stats
contingency_table=pd.crosstab(df['Gender'],df['Designation'])
chi2_stats,p_val,dof,expected=stats.chi2_contingency(contingency_table)
alpha=0.05
if p_val< alpha:
    print("There is a significant relationship between gender and specialization preference.")
else:
     print("There is no significant relationship between gender and specialization preference.")
```

### Conclusion

<font size="4">Based on the analysis, the claim suggesting a salary range of 2.5-3 lakhs for Computer Science Engineering graduates in specific job roles is not supported by the data. Additionally, the statistical test results indicate that there is no significant relationship between gender and specialization preference among  graduates, as per the AMCAT dataset.
